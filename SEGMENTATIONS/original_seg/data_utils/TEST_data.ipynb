```python
import numpy as np
import pandas as pd
import os
import pydicom
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
xrayfi = "../../data/segmentation/data/xray/"
segfi = "../../data/segmentation/data/seg/"
```

```python
all_xray_files = [i[:-4] for i in os.listdir(xrayfi) if i.lower().endswith("dcm")]
all_seg_files = [i[:-7] for i in os.listdir(segfi) if i.lower().endswith("nii.gz")]
all_files = list(set(all_xray_files) & set(all_seg_files))
```

```python
for i in all_files:
    ds = pydicom.dcmread(xrayfi + i + ".dcm")
    xrayimg = ds.pixel_array
    
    segimg = nib.load(segfi + i + ".nii.gz").get_fdata().transpose(1,0,2).squeeze()

    print(xrayimg.shape, segimg.shape)

    plt.imshow(xrayimg, cmap="gray")
    plt.axis("off")
    plt.show()

    plt.imshow(segimg, cmap="gray")
    plt.axis("off")
    plt.show()
```

```python
i = "9442268L"

img = nib.load(segfi + i + ".nii.gz")
data = img.get_fdata()

print("Original shape:", data.shape)

# Remove first channel (index 0)
data_new = data[:, :, 1]

# If you want to keep it 3D with single channel:
data_new = data_new[:, :, np.newaxis]

print("New shape:", data_new.shape)

# Save new NIfTI
new_img = nib.Nifti1Image(data_new, img.affine, img.header)
nib.save(new_img, segfi + i + ".nii.gz")
```
